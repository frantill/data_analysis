## Pearson's r Correlation

```python
def correlation(x, y):
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    x_std = (x - x.mean())/x.std(ddof=0) # data standardization
    y_std = (y - y.mean())/y.std(ddof=0) # take uncorrected std passinf ddof=0
    r = (x_std*y_std).mean()
    return r

entries = subway_df['ENTRIESn_hourly']
cum_entries = subway_df['ENTRIESn']
rain = subway_df['meanprecipi']
temp = subway_df['meantempi']

print("Entries-rain correlation: %f" % correlation(entries, rain))
print("Entries-temp correlation: %f" % correlation(entries, temp))
print("Rain-temp correlation: %f" % correlation(rain, temp))

print("Entries-cumulative entries correlation: %f" % correlation(entries, cum_entries))
```
